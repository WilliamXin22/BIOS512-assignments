Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
library('scales')
```

```r
options(repr.plot.width = 5, repr.plot.height = 6)
p = ggplot(penguins, aes(x = body_mass_g)) +
    geom_histogram(bins = 37, color = 'steelblue', fill = 'steelblue', alpha = 0.8) + 
    scale_y_continuous(labels = comma_format()) + 
    scale_x_continuous(labels = comma_format()) +
    facet_wrap(~year, ncol = 1) + 
    theme_minimal() + 
    theme(strip.text=element_text(hjust = 0.925,size = 8,color = 'grey50',face = 'bold')) +
    labs(title = 'Histograms of penguin body mass for measurements taken from 2007-2009', x = 'Body mass in grams', y = 'Numbers of penguins') +
    theme(plot.title = element_text(size = 8, hjust = 0.01), 
          axis.text.x = element_text(size = 7),
          axis.text.y = element_text(size = 8),
          axis.title.x = element_text(size = 10), 
          axis.title.y = element_text(size = 10))

p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
p = ggplot(penguins, aes(x = body_mass_g, y = sex)) + 
    geom_point(position = position_jitter(height = 0.25, width = 0), alpha = 0.4) +
    facet_wrap(~species, ncol = 1) +
    scale_x_continuous(labels = comma_format(),breaks = seq(3500,5500,by = 1000)) +
    theme_minimal() +
    theme(strip.text=element_text(hjust = 0.1,size = 8,color = 'grey50',face = 'bold')) +
    labs(x = 'Body mass in grams', y = '')
p
```
